## Importando as Bibliotecas 

```python
import pandas as pd 
import numpy as np 
import datetime
```

## Ler o Dataset de Custo

```python
data = pd.read_csv("Custos.csv", encoding= 'windows-1258', sep=',')
data
```

## Adicionando a coluna Código para manipular Strings do dataset de Custo

```python
data['Codigo'] = data['Produto'].str[2:]
data
```

## Renomeando o nome da coluna Produto para Produto 1

```python
data['Produto 1'] = data['Produto'].str[:2]
data
```

## Deletando a coluna Produto

```python
data = data.drop(columns=['Produto'])
data.head()
```

## Inserindo as Colunas Dia, Mês e Ano 

```python
data["Dia"] = pd.DatetimeIndex(data["Data"]).day
data["Mês"] = pd.DatetimeIndex(data["Data"]).month
data["Ano"] = pd.DatetimeIndex(data["Data"]).year

data
```

## Deletando a coluna Data 

```python
data = data.drop(columns=['Data'])
```

```python
data.head()
```

## Calculo de Vendas e Custo da Empresa X

```python
total_vendas = data["Valor de Venda"].sum()
total_custos = data["Preço Custo"].sum()
media_vendas = data["Valor de Venda"].mean() 
media_custos = data["Preço Custo"].mean()
faturamento = total_vendas * total_custos
lucro = total_vendas - total_custos

analise_v_c = pd.DataFrame({"Total de Vendas" : [total_vendas], 
                            "Total de Custos" : [total_custos], 
                            "Média de Vendas" : [media_vendas], 
                            "Média de Custos" : [media_custos],
                            "Faturamento" : [faturamento], 
                            "Lucro" : [lucro]})

analise_v_c["Total de Vendas"] = analise_v_c["Total de Vendas"].map("R$ {:,.2f}".format)
analise_v_c["Total de Custos"] = analise_v_c["Total de Custos"].map("R$ {:,.2f}".format)
analise_v_c["Média de Vendas"] = analise_v_c["Média de Vendas"].map("R$ {:,.2f}".format)
analise_v_c["Média de Custos"] = analise_v_c["Média de Custos"].map("R$ {:,.2f}".format)
analise_v_c["Faturamento"] = analise_v_c["Faturamento"].map("R$ {:,.2f}".format)
analise_v_c["Lucro"] = analise_v_c["Lucro"].map("R$ {:,.2f}".format)

analise_v_c
```

## Duraçăo Venda Telefone (mins) e Tempo Preparaçăo (mins) por Produto 

```python
duracao_venda_t = data.groupby(["Produto 1"]).sum()["Duraçăo Venda Telefone (mins)"].rename("Duração de Vendas por Telefone")
tempo_preparacao_min = data.groupby(["Produto 1"]).sum()["Tempo Preparaçăo (mins)"].rename("Tempo de Preparação")

analise_p_t = pd.DataFrame({"Duração de Vendas por Telefone" : duracao_venda_t, 
                            "Tempo de Preparação" : tempo_preparacao_min})

analise_p_t
```

## Analise Geral de Cálculos de Mês por Valor de Venda

```python
total_valor_venda_mes = data.groupby(["Mês"]).sum()["Valor de Venda"].rename("Total")
media_valor_venda_mes = data.groupby(["Mês"]).mean()["Valor de Venda"].rename("Média")
desvio = data.groupby(["Mês"]).std()["Valor de Venda"].rename("Desvio Padrão")
variancia = data.groupby(["Mês"]).var()["Valor de Venda"].rename("Variância")
quartil = data.groupby(["Mês"]).quantile(0.50)["Valor de Venda"].rename("Quartil")


analise_mes = pd.DataFrame({"Total" : total_valor_venda_mes, 
                            "Média" : media_valor_venda_mes, 
                            "Desvio Padrão" : desvio, 
                            "Variância" : variancia, 
                            "Quartil" : quartil})

analise_mes["Total"] = analise_mes["Total"].map("R$ {:,.2f}".format)
analise_mes["Média"] = analise_mes["Média"].map("R$ {:,.2f}".format)
analise_mes["Desvio Padrão"] = analise_mes["Desvio Padrão"].map("R$ {:,.2f}".format)
analise_mes["Variância"] = analise_mes["Variância"].map("R$ {:,.2f}".format)
analise_mes["Quartil"] = analise_mes["Quartil"].map("R$ {:,.2f}".format)
analise_mes
```

## Analise Geral de Cálculos de Mês por Preço de Custo 

```python
total_preco_custo_mes = data.groupby(["Mês"]).sum()["Valor de Venda"].rename("Total")
media_preco_custo_mes = data.groupby(["Mês"]).mean()["Valor de Venda"].rename("Média")
desvio = data.groupby(["Mês"]).std()["Valor de Venda"].rename("Desvio Padrão")
variancia = data.groupby(["Mês"]).var()["Valor de Venda"].rename("Variância")
quartil = data.groupby(["Mês"]).quantile(0.50)["Valor de Venda"].rename("Quartil")


analise_mes = pd.DataFrame({"Total" : total_preco_custo_mes, 
                            "Média" : media_preco_custo_mes, 
                            "Desvio Padrão" : desvio, 
                            "Variância" : variancia, 
                            "Quartil" : quartil})

analise_mes["Total"] = analise_mes["Total"].map("R$ {:,.2f}".format)
analise_mes["Média"] = analise_mes["Média"].map("R$ {:,.2f}".format)
analise_mes["Desvio Padrão"] = analise_mes["Desvio Padrão"].map("R$ {:,.2f}".format)
analise_mes["Variância"] = analise_mes["Variância"].map("R$ {:,.2f}".format)
analise_mes["Quartil"] = analise_mes["Quartil"].map("R$ {:,.2f}".format)
analise_mes
```

## Faturamento e Lucro por Mês 

```python
faturamento = total_valor_venda_mes * total_preco_custo_mes
lucro = total_valor_venda_mes + total_preco_custo_mes

analise_mes = pd.DataFrame({"Faturamento por Mês" : faturamento,
                            "Lucro": lucro})

analise_mes["Faturamento por Mês"] = analise_mes["Faturamento por Mês"].map("R$ {:,.2f}".format)
analise_mes["Lucro"] = analise_mes["Lucro"].map("R$ {:,.2f}".format)

analise_mes
```
